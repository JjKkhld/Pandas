#Stworzenie danych

```python
nabrzeza = pd.read_excel('/content/drive/MyDrive/DataFrames/Dane przeładunkowe.xlsx',
                     sheet_name='Sheet1')

przeladunki = pd.read_excel('/content/drive/MyDrive/DataFrames/Dane przeładunkowe.xlsx',
                     sheet_name='2023')
```

```python
nabrzeza.rename(columns=lambda x: x.strip(), inplace=True)
```

```python
nabrzeza = nabrzeza.drop('Nr', axis=1)
```

```python
nabrzeza['Nazwa'] = nabrzeza['Nazwa'].apply(lambda x: x.replace('Nabrzeże', ''))
nabrzeza['Nazwa'] = nabrzeza['Nazwa'].apply(lambda x: x.strip())

# cleaninng 'name'
nabrzeza['Nazwa'] = nabrzeza['Nazwa'].apply(lambda x: x.replace('Wałbrzyske', 'Wałbrzyskie'))
```

```python
przeladunki = przeladunki.drop(index=[58,59], axis=0)
```

```python
przeladunki.rename(columns=lambda x: x.strip(), inplace=True)
```

```python
str_to_drop = ['Podwykonawca', 'Szczecin']

przeladunki = przeladunki[przeladunki['N a z w a\xa0 n a b r z e ż a'].str.contains('Podwykonawca|Szczecin') == False]
```

```python
przeladunki = przeladunki.drop(['Lp.', 'zmiana'], axis=1)
przeladunki = przeladunki.rename(columns={'N a z w a\xa0 n a b r z e ż a':'Nazwa'})
przeladunki['Nazwa'] = przeladunki['Nazwa'].apply(lambda x: x.strip())

#cleaning 'Nazwa'
przeladunki['Nazwa'] = przeladunki['Nazwa'].apply(lambda x: x.replace('Parnica', 'Parnickie'))
przeladunki['Nazwa'] = przeladunki['Nazwa'].apply(lambda x: x.replace('Parnica', 'Parnickie'))
przeladunki['Nazwa'] = przeladunki['Nazwa'].apply(lambda x: x.replace('CPN 3 Sz-n', 'CPN-3'))
przeladunki['Nazwa'] = przeladunki['Nazwa'].apply(lambda x: x.replace('Drzetowskie', 'Drzetowskie N'))
przeladunki['Nazwa'] = przeladunki['Nazwa'].apply(lambda x: x.replace('Farbolak', 'Farblak'))
przeladunki['Nazwa'] = przeladunki['Nazwa'].apply(lambda x: x.replace('Kra', 'Krakowskie'))
przeladunki['Nazwa'] = przeladunki['Nazwa'].apply(lambda x: x.replace('Snop', 'SNOP N (1)'))
przeladunki['Nazwa'] = przeladunki['Nazwa'].apply(lambda x: x.replace('Wulkan', 'Wulkan N'))
przeladunki = przeladunki.drop(index=26).reset_index().drop('index', axis=1)
```

```python
# deleting dupl

cols_to_proc = ['Niemieckie', 'Rumuńskie', 'Słowackie']

for i in cols_to_proc:
  indexes_to_proc = przeladunki['Nazwa'].loc[przeladunki['Nazwa'] == i].index
  data_to_proc = przeladunki.iloc[[indexes_to_proc[0], indexes_to_proc[1]]]
  sum = data_to_proc.sum()
  sum.update(pd.Series([i], index=['Nazwa']))
  przeladunki.loc[len(przeladunki)] = sum
  przeladunki = przeladunki.drop(index=[indexes_to_proc[0], indexes_to_proc[1]]).reset_index().drop('index', axis=1)
```

```python
cols = przeladunki.columns[1:]
przeladunki[cols] = przeladunki[cols].apply(lambda x: round(x / 1000, 2))
```

```python
dane_przel_nadrzeza = nabrzeza.merge(przeladunki, how='left', on='Nazwa').fillna(0)
```

```python
dane_przel_nadrzeza['Nazwa'] = dane_przel_nadrzeza['Nazwa'].apply(lambda x: x.strip())
```

```python
dane_przel_nadrzeza = dane_przel_nadrzeza.set_index('Nazwa')
```

```python
all_taking = list(przeladunki['Nazwa'])

nabrzeza_z_przel = dane_przel_nadrzeza.loc[all_taking].reset_index().sort_values(by=['O g ó ł e m'], ascending=True)
```

```python
nabrzeza_z_przel['Ładunki_masowe'] = nabrzeza_z_przel['O g ó ł e m'] - nabrzeza_z_przel['Drobnica']
```

```python
search_bulk = ['Parnickie', 'Noteckie', 'Górnośląskie', 'Poznańskie',
          'Bydgoskie', 'Gliwickie', 'Bytomskie', 'Wałbrzyskie',
          'Chorzowskie', 'Katowickie', 'Regalica']
search_general = ['Rosyjskie', 'Greckie', 'Rumuńskie', 'Węgierskie', 'Polskie', 'Angielskie',
          'Belgijskie', 'Holenderskie', 'Zbożowe', 'Słowackie',
          'Czeskie', 'Fińskie']

nabrzeza_z_przel['Strefa'] = nabrzeza_z_przel['Nazwa'].apply(lambda x: 'BulkCargo' if x in search_bulk else 'GenCargo' if x in search_general else 'empty')
```
